### Gender 
It would help our analysis if we know the ratio of female to male and the influence on students performance. Gender column will  be helpful in doing this

```python
gender_size = data.gender.value_counts()         ## No. of male and female
gender_size
```

```python
plt.figure(figsize=(12, 6))
plt.title('Percentage of Male and Female')
plt.pie(gender_size, labels=gender_size.index, autopct='%1.1f%%', startangle=270);
```

It turns out that about 48.2% are male and 51.8% are female

```python
## Influence of gender on students performance
```

```python
plt.figure(figsize=(14, 8))
plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course', y='math score', data=data, hue='gender', palette='viridis', 
            edgecolor='black', **{'hatch':"/", 'linewidth':2})
           
plt.title('MATH SCORES')

plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course', y='reading score', data=data, hue='gender', palette='viridis',   
            edgecolor='black', **{'hatch':"|", 'linewidth':2, 'linestyle':':'})
        
plt.title('READING SCORES')

plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course', y='writing score', data=data, hue='gender', palette='viridis',  
            edgecolor='black', **{'hatch':'-', 'linewidth':2, 'linestyle':'--'})
         
plt.title('WRITING SCORES')
plt.show()
```

It can be seen that male have higher scores in math test while in reading and writing test, female have higher scores

###  Asking and Answering Questions

#### Q.2 What affects the students performance the most between the lunch taken and test prep course?

To compute the correlation for these columns, they must first be converted into numeric columns as they contain categorical column

```python
## Checking the correlation between the two columns
```

#### Lunch


```python
lunch_values = {'standard': 0, 'free/reduced': 1}
lunch_numeric = data.lunch.map(lunch_values)
data['total score'].corr(lunch_numeric)
```

#### Test preparation score

```python
test_prep_values = {'completed': 0, 'none': 1}
test_prep_numeric = data['test preparation course'].map(test_prep_values)
data['total score'].corr(test_prep_numeric)
```

How students prepared for their test is more important that the lunch they're taking, test preparation course greatly affect students performance than the lunch, i.e it's possible to have a free/reduced lunch and still pass excellently

#### Q 2. How many students completed their test prep course, had standard lunch and also passed excellently?

```python
data[(data['math score'] > 90) & (data['reading score'] > 90) & (data['writing score'] > 90) & 
     (data['test preparation course'] == 'completed') & (data['lunch'] == 'standard')]
```

It can be seen that only 11 students completed their test preparation course and also had a standard lunch, of course they topped the class in all the test scores
